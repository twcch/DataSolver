```python
import os
import pandas as pd
```

```python
src_folder = "/Users/twcch/Documents/Drive/Data/新光人壽資料/SPSS/"
desc_folder = "/Users/twcch/Documents/Drive/Data/新光人壽資料/Data"
file_name = "XCLSD(987149)"
```

```python
data = pd.read_spss(os.path.join(src_folder, file_name + ".sav"))
data
```

```python
data.columns
```

```python
def transfer_type(df, columns, type="str"):
    for column in columns:
        df[column] = df[column].astype(type).map(lambda x : x.split(".")[0])
    return df

data = transfer_type(data, ["claim_rece_no", "hosp_code", "visit_type", "claim_date_start", "claim_date_end", "seq_no", "surgery_code_ins", "surgery_code", "surgery_date", "major_ind"])
```

```python
data
```

```python
data.to_csv(os.path.join(desc_folder, file_name + ".csv"), index=False)
```
