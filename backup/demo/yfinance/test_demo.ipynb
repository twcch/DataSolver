```python
import yfinance as yf
import datetime as dt

start = dt.datetime(2015, 1, 1)
end = dt.datetime(2024, 12, 31)

data = yf.download(tickers="GOOG", start=start, end=end)

data
```

```python
type(data.index)
```

## DatetimeIndex

```python
# 取出每筆資料的週數
data["week"] = data.index.isocalendar().week
data
```

## 計算每週交易量的總和

```python
data.groupby("week").Volume.sum()
```

## resample

```python
# 調整週期重新取樣
# 股票 10 日均線，需要 10 日 10 日的重新計算
ten = data.resample("10D").mean()  # 代表 10 日
ten  # 會得到每 10 天平均的結果
```
